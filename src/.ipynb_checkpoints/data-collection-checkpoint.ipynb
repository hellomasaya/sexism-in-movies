# 1. Collecting Movie scripts using web scraping

```python
import re
import os
import codecs
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
def get_movie_info():
    link_all_scripts = 'https://imsdb.com/all-scripts.html'
    response_all_scripts = requests.get(link_all_scripts)
    soup = BeautifulSoup(response_all_scripts.text, 'html.parser')
    find_tables = soup.findAll('td', valign='top')
    all_movies = find_tables[2].findAll('a')
    movies = [(movie_info.string, \
        movie_info["href"], \
        re.split("[,.]",movie_info.string)[0].replace(' ', '_'))
      for movie_info in all_movies]
    return movies
```

```python
def get_scripts(movie, driver):
    title, link_to_movie_page, movie_title = movie

    full_html_link = u'http://www.imsdb.com' + link_to_movie_page
    response_script = requests.get(full_html_link)
    soup = BeautifulSoup(response_script.text, 'html.parser')

    list_links = soup.findAll('table', "script-details")[0].findAll('a')
    genre = []
    writer = []
    script = ''
    for link in list_links:
        href = link['href']
        if href[0:7]== "/writer":
            writer.append(link.get_text())
        if href[0:7]== "/genre/":
            genre.append(link.get_text())
        if href[0:9]== "/scripts/":
            script = href

    if script == '' or script[-5:] != '.html':
        return;
        path_to_directory = '../data/'
        pdf_logging_filename = path_to_directory + 'movies_pdf_script.csv'
        with open(pdf_logging_filename, 'a') as f:
            new_row = title + '\n'
            f.write(new_row)

    else:

        full_script_url =  u'http://www.imsdb.com' + script
        driver.get(full_script_url)
        page_text = driver.page_source
        soup = BeautifulSoup(page_text, 'html.parser')

        if len(soup.findAll('td', "scrtext"))!=1:
            error_file_name = '../data/scraping_error.csv'
            with open(error_file_name, 'a') as error_file:
                new_row = title + '\n'
                error_file.write( new_row )

        else:
            path_to_directory = '../data/scripts/'
            filename = path_to_directory + movie_title + '.txt'
            text = soup.findAll('td', "scrtext")[0].get_text()
            with codecs.open(filename, "w",encoding='ascii', errors='ignore') as f:
                f.write(text)

            path_to_directory = '../data/'
            success_filename = path_to_directory + 'successful_files2.csv'
            new_row = title + ';' + str(genre) + ';' + str(writer) + ';' + movie_title + ';' + filename + '\n'
            with open(success_filename, 'a') as f:
                f.write(new_row)
```

```python
if not os.path.exists('../data'):
    os.mkdir('../data')
if not os.path.exists('../data/scripts'):
    os.mkdir('../data/scripts')
movies = get_movie_info()
driver_path = "/Users/harshitasharma/Downloads/chromedriver"
driver = webdriver.Chrome(driver_path)
for i,movie in enumerate(movies):
    print(i, "/", len(movies), " ,", (i/len(movies))*100, "%\n")    
    get_scripts(movie, driver)
```

# 2. Years of movie release

```python
import pandas as pd
import imdb
# 1167, 1178, 776, 823
```

```python
movieinfo = pd.read_csv('../data/successful_files.csv', delimiter="\t")
# movieinfo = movieinfo.sort_values(by='Movie', ascending=True, na_position='first')
```

```python
movieinfo.head(7)
```

```python
def get_year_of_release(movies):
    ia = imdb.IMDb()
    years = []
    for ind in movies.index:
        mov = movies['Movie'][ind]
        try:
            search = ia.search_movie(mov) 
            y = search[0]['year']
            years.append(y)
            print(ind, "/", len(movies), ":", y)
        except:
            print(ind, " ", mov, "NOT AVAILABLE")
            years.append(" ")
    return years
```

```python
years = get_year_of_release(movieinfo)
```

```python
movieinfo['Year'] = years
```
