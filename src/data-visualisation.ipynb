# Data visualisation
## 1. By Year of release

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#year, genre
movieinfo = pd.read_csv('../data/final_movies_list.csv', delimiter="\t")
years = pd.unique(movieinfo['Year'])
```

```python
print(np.sort(years))
```

```python
# plt.figure(figsize=(17, 3))
sns.set(font_scale=1)
# sns.set(rc={'figure.figsize':(17,3)})
sns.catplot(x="Year", kind="count", palette="ch:.25", data=movieinfo, height=15.27, aspect=11.7/5.27)
```

```python
yearly = movieinfo['Year'].value_counts()
yearly.head(10)
```

```python
y = movieinfo['Year']
```

```python
for i, j in enumerate(y):
    if j < 1930:
        y[i] = 1930
    elif j < 1945:
        y[i] = 1945
    elif j < 1960:
        y[i] = 1960
    elif j < 1975:
        y[i] = 1975
    elif j < 1990:
        y[i] = 1990
    elif j < 2005:
        y[i] = 2005
    elif j < 2020:
        y[i] = 2020
    elif j < 2022:
        y[i] = 2022
print(y)
```

```python
for i, j in enumerate(y):
    if j == 1930:
        y[i] = '1915-30'
    elif j == 1945:
        y[i] = '1930-45'
    elif j == 1960:
        y[i] = '1945-60'
    elif int(j) == 1975:
        y[i] = '1960-75'
    elif int(j) == 1990:
        y[i] = '1975-90'
    elif int(j) == 2005:
        y[i] = '1990-2005'
    elif int(j) == 2020:
        y[i] = '2005-20'
    elif j == 2022:
        y[i] = '2020-22'
```

```python
print(y)
```

```python
yearly3 = pd.DataFrame(y, index=None, columns = ['Year'])
yearly3 = yearly3.sort_values(by = ['Year'])
yearly3.head(10)
```

```python
sns.set(font_scale=4)
sns.catplot(x="Year", kind="count", palette="ch:.25", data=yearly3, height=15.27, aspect=11.7/5.27)
```

## 2. By genre

```python
genre = pd.DataFrame(movieinfo['Genre'])
```

```python
genre.head(10)
```

```python
unique_genre = []
for ind in genre.index:
    g = genre['Genre'][ind]
    g = g.replace('[', "").replace(']', "").replace('\'', "")
    g = g.split(", ")
    for i in g:
        if i not in unique_genre:
            unique_genre.append(i)
```

```python
print(unique_genre)
```

```python
count = []
for i, j in enumerate(unique_genre):
    temp = [j, 0]
    count.append(temp)
    
count = np.asarray(count)
# print(count[:, 0] )
# index = np.where(count[:, 0] == 'History')
# print(index[0][0])
for ind in genre.index:
    g = genre['Genre'][ind]
    g = g.replace('[', "").replace(']', "").replace('\'', "")
    g = g.split(", ")
    for i in g:
        index = np.where(count[:, 0] == i)
        count[index[0][0]][1] = int(count[index[0][0]][1]) + 1
```

```python
print(count)
```

```python
bygenre = pd.DataFrame(count, columns=['Genre', 'Count'])
bygenre['Count'] = bygenre['Count'].astype(int)
```

```python
bygenre.head(10)
```

```python
sns.set(font_scale=2)
sns.catplot(x="Genre", y="Count", kind = "bar", palette="ch:.25", data=bygenre, height=15.27, aspect=11.7/5.27)
```
